```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread(r'C:\Users\ASUS\Desktop\python_analysis\photo2.PNG')

b, g, r = cv2.split(image)

hist_b = cv2.calcHist([b], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])

plt.figure(figsize=(12, 6))

plt.subplot(3, 2, 1)
plt.imshow(b, cmap='Blues')
plt.title('Mavi Kanal')
plt.colorbar()
plt.subplot(3, 2, 2)
plt.bar(range(256), hist_b.flatten(), color='b', width=1)
plt.title('Blue Channel Histogram')
plt.xlim([0, 256])

plt.subplot(3, 2, 3)
plt.imshow(g, cmap='Greens')
plt.title('Yeşil Kanal')
plt.colorbar()
plt.subplot(3, 2, 4)
plt.bar(range(256), hist_g.flatten(), color='g', width=1)
plt.title('Green Channel Histogram')
plt.xlim([0, 256])

plt.subplot(3, 2, 5)
plt.imshow(r, cmap='Reds')
plt.title('Kırmızı Kanal')
plt.colorbar()
plt.subplot(3, 2, 6)
plt.bar(range(256), hist_r.flatten(), color='r', width=1)
plt.title('Red Channel Histogram')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread(r'C:\Users\ASUS\Desktop\python_analysis\photo2.PNG', cv2.IMREAD_GRAYSCALE)

threshold_value = 127

ret, binary_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('B-W Photo')

plt.subplot(1, 2, 2)
plt.imshow(binary_image, cmap='binary')
plt.title('Binary Photo')

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread(r'C:\Users\ASUS\Desktop\python_analysis\photo2.PNG')

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

threshold_value1 = 127
threshold_value2 = 200

ret1, binary_image1 = cv2.threshold(gray_image, threshold_value1, 255, cv2.THRESH_BINARY)
ret2, binary_image2 = cv2.threshold(gray_image, threshold_value2, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Gray Tones')

plt.subplot(1, 3, 2)
plt.imshow(binary_image1, cmap='binary')
plt.title('Eşik Değeri: {}'.format(threshold_value1))

plt.subplot(1, 3, 3)
plt.imshow(binary_image2, cmap='binary')
plt.title('Eşik Değeri: {}'.format(threshold_value2))

plt.show()
```
